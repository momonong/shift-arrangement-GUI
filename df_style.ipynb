```python
import pandas as pd
```

```python
def font_color(val):
    color = 'yellow' if val=='公休' else 'red'
    return f'color: {color}' 

def background_color(val):
    color = 'grey' if val=='84-7' else ''
    return f'backround-color: {color}'
```

```python
df = pd.read_csv('csv_files/_ALL_shift.csv')
df = df.head(5)
df
```

```python
new_df = df.style.applymap(font_color, subset=pd.IndexSlice[:, ['7']])
new_df
```

```python
new_df2 = df.style.applymap(background_color, subset=pd.IndexSlice[:, ['16']])
new_df2
```
